```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('energydata_complete.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
renamed_columns = {'T2': 'Temperature Inside', 'T6': 'Temperature Outside'}
```

```python
df = df.rename(columns=renamed_columns)
```

```python
df.head()
```

```python
reg_df = df[['Temperature Inside', 'Temperature Outside']].sample(100, random_state=2)
```

```python
#regression plot
sns.regplot(x='Temperature Inside', y='Temperature Outside', data=reg_df)
```

```python
#preprocessing
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df = df.drop(columns=['date'])
```

```python
normalized_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
df.head()
```

```python
normalized_df.head()
```

```python
features_df = normalized_df[['Temperature Inside']]
```

```python
features_df.head()
```

```python
target = normalized_df['Temperature Outside']
```

```python
target.head()
```

```python
#modelling
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, target, test_size = 0.3, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train, y_train)
```

```python
pred_values = linear_model.predict(x_test)
```

```python
pred_values
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score = r2_score(y_test, pred_values)
```

```python
round(r2_score, 3)
```
